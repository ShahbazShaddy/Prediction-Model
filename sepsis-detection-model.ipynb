```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, confusion_matrix, precision_recall_curve
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/kaggle/input/sepsis-prediction/Dataset.csv')  # Replace with your file path if necessary
print(df.head())  # Inspect the first few rows
```

```python
df.fillna(df.median(), inplace=True)  # Filling missing values with median
```

```python
# Define X (features) and y (target)
X = df.drop(['SepsisLabel', 'Patient_ID'], axis=1)  # Features
y = df['SepsisLabel']  # Target

# Check data types to make sure they're all numeric
print(X.dtypes)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

# Accuracy and classification report
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
