```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
authors = [
    "Qinhao Wu et al.",
    "Schertz et al.",
    "Al-Mualemi and Lu",
    "Zhang et al.",
    "Rosnati and Fortuin",
    "Abromaviius et al.",
    "Proposed Method"
]

accuracy = [0.839, 0.79, 0.938, None, None, 0.934, 0.956]
auroc = [0.89, None, None, 0.892, 0.660, 0.307, 0.899]

# Filter data to remove N/A (None) values
filtered_accuracy = [(authors[i], val) for i, val in enumerate(accuracy) if val is not None]
filtered_auroc = [(authors[i], val) for i, val in enumerate(auroc) if val is not None]

# Separate labels and values after filtering
accuracy_authors, accuracy_values = zip(*filtered_accuracy)
auroc_authors, auroc_values = zip(*filtered_auroc)

# Function to add values on top of bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width() / 2, height + 0.02,
            f'{height:.3f}', ha='center', va='bottom', fontsize=10
        )

# Plotting Accuracy Comparison
plt.figure(figsize=(12, 7))
x_accuracy = np.arange(len(accuracy_authors))
bars1 = plt.bar(x_accuracy, accuracy_values, color='skyblue', alpha=0.8, edgecolor='black')
add_value_labels(bars1)
plt.xticks(x_accuracy, accuracy_authors, rotation=45, ha='right', fontsize=13)
plt.title('Accuracy Comparison', fontsize=18, fontweight='bold')
plt.xlabel('Authors', fontsize=18)
plt.ylabel('Accuracy', fontsize=18)
plt.ylim(0, 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('../graph/accuracy_comparison.pdf')
plt.show()

# Plotting AUROC Comparison
plt.figure(figsize=(12, 7))
x_auroc = np.arange(len(auroc_authors))
bars2 = plt.bar(x_auroc, auroc_values, color='orange', alpha=0.8, edgecolor='black')
add_value_labels(bars2)
plt.xticks(x_auroc, auroc_authors, rotation=45, ha='right', fontsize=13)
plt.title('AUROC Comparison', fontsize=18, fontweight='bold')
plt.xlabel('Authors', fontsize=18)
plt.ylabel('AUROC', fontsize=18)
plt.ylim(0, 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('../graph/auroc_comparison.pdf')
plt.show()
```
